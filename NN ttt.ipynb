```python
class strategy0:
    def __init__(self,w,b):
        self.weights=w
        self.biases=b
    def prob(self,input_values):
        exponential=np.matmul(self.weights,input_values.reshape(9,-1))+self.biases
        return exponential
    def move(self,field):
        x=field.reshape(9,-1)
        p=self.prob(field)
        return p==(np.max(p[x==0]))
    def decision(self,input_values):
        return np.argmax(self.prob(input_values))
    def dec_oh(self,input_values):
        position=self.decision(input_values)
        return np.eye(position+1)[position]


class history:
    def __init__(self,w,b):
        self.weights=w
        self.biases=b


def sigmoid(var):
    return 1/(1+np.exp(-var))


    
def game_over_q(field):
    plus=np.array([1,1,1])
    minus=np.array([-1,-1,-1])
    def indicator(vec):
        return ((field[0]==vec).all() or (field[1]==vec).all() or (field[2]==vec).all() 
            or (field.T[0]==vec).all() or (field.T[1]==vec).all() or (field.T[2]==vec).all() 
            or (np.diag(field)==vec).all() or (np.diag(np.fliplr(field))==vec).all())    
    if indicator(plus):
        return np.array([1,-1])
    elif indicator(minus):
        return np.array([-1,1])
    else:
        return np.array([0,0])
    


def game(strat1,strat2):
    field=np.zeros((3,3))
    counter=0
    while (not (game_over_q(field)).any()) and counter<10:
        #print('Step 1: \n')
        field=field+strat1.move(field).reshape(3,3)
        counter=counter+1
        #print(field)
        if (game_over_q(field)).any() or counter>8:
            break
        field=field-strat2.move(field).reshape(3,3)
        counter=counter+1
        #print(field)
    return game_over_q(field)

def breading_1(strat1,strat2,rel_frac):
    mutation_rate=0.02
    w=strat1.weights*(1+(2*np.random.random()-1)*mutation_rate)*rel_frac/(1+rel_frac)+strat2.weights*(1+(2*np.random.random()-1)*mutation_rate)/(1+rel_frac)
    b=strat1.biases*(1+(2*np.random.random()-1)*mutation_rate)*rel_frac/(1+rel_frac)+strat2.biases*(1+(2*np.random.random()-1)*mutation_rate)/(1+rel_frac)
    return strategy0(w,b)

def new_generation(players):
    return
```

```python
np.random.seed(0)
n_best=4
n_random=15
num_players=np.int(n_best*(n_best+1)/2)+n_random
players=np.array([strategy0(np.random.random((9,9)),np.random.random((9,1))) for i in range(num_players)])
hist=history(np.array([]),np.array([]))
generation=0
num_players
```

```python
for iteration in range(10**4):
    scores=np.zeros(num_players)
    for i in range(num_players):
        for j in range(num_players):
            #clear_output(wait=True)
            match=game(players[i],players[j])
            #print('Currently playing: Player',i,'vs.','Player',j,'Result :',match)
            scores[i]=scores[i]+match[0]
            scores[j]=scores[j]+match[1]
    print('Generation: ',generation,' Score: ',np.flip(np.sort(scores),0))
    
    if (scores==0).all():
        break
    else:    
        qual_ind=np.argsort(scores)[-n_best:]
        players_qual=players[qual_ind]
        scores_qual=scores[qual_ind]
        rel_scores=scores_qual-np.min(scores)+1
        fractions=rel_scores/rel_scores.sum()

        current_best=players[np.argmax(scores)]

        #hist.weights=np.append(hist.weights,current_best.weights)
        
        #fractions[i]/fractions[j]
        
        players_new=np.array([])
        for i in range(n_best):
            for j in range(i,n_best):
                players_new=np.append(players_new,breading_1(players_qual[i],players_qual[j],fractions[i]/fractions[j]))
        
        players_random=np.array([strategy0(np.random.random((9,9)),np.random.random((9,1))) for i in range(n_random)])
        
        players=np.append(players_new,players_random)
        generation=generation+1
```
